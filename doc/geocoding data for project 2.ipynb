```r
library(ggmap)

ems2016  <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_ems.xlsx'))
ems2017  <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2017_ems.xlsx'))
ems2018  <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2018_ems.xlsx'))

hs2016 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_hs.xlsx'))
hs2017 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_hs.xlsx'))
hs2018 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_hs.xlsx'))

pk2016 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_pk.xlsx'))
pk2017 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_pk.xlsx'))
pk2018 <- data.frame(readxl::read_xlsx('/Users/marshall/Documents/GitHub/fall2019-proj2--sec2-grp11/data/2016_pk.xlsx'))
```

# ems and hs school names that need lat and lon

```r
ems  <- Reduce(union, list(ems2018[,2],ems2017[,2],ems2016[,2]))
hs  <- Reduce(union, list(hs2018[,2],hs2017[,2],hs2016[,2]))
```

# pk full address

```r
pk16  <- paste(pk2016[2:nrow(pk2016),'Address.1'],',',pk2016[2:nrow(pk2016),'Address.2'])
pk17  <- paste(pk2017[2:nrow(pk2017),'Address.1'],',',pk2017[2:nrow(pk2017),'Address.2'])
pk18  <- paste(pk2018[2:nrow(pk2018),'Address.1'],',',pk2018[2:nrow(pk2018),'Address.2'])

pk  <- Reduce(union, list(pk16,pk17,pk18))
```

# Geocoding

```r
emscoord = data.frame(geocode(ems,output = 'latlon',source = 'google'))
```

```r
ems_latlon  <- cbind(data.frame((ems), stringsAsFactors=FALSE),emscoord)
```

```r
hscoord = data.frame(geocode(hs,output = 'latlon',source = 'google'))
```

```r
hs_latlon  <- cbind(data.frame((hs), stringsAsFactors=FALSE),hscoord)
```

```r
pkcoord = data.frame(geocode(pk,output = 'latlon',source = 'google'))
```

```r
pk_latlon  <- cbind(data.frame((pk), stringsAsFactors=FALSE),pkcoord)
```

```r
saveRDS(ems_latlon, file = "ems_latlon.rds")
saveRDS(hs_latlon,file = 'hs_latlon.rds')
saveRDS(pk_latlon,file = 'pk_latlon.rds')
```
