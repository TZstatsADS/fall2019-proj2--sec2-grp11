```r
library(stringr)
```

```r
ems  <- readRDS('ems_latlon.rds')
hs  <- readRDS('hs_latlon.rds')
```

# Reverse Geocoding for ems

```r
zip = c()
for (i in 1:nrow(ems)){
    res = revgeocode(c(ems[i,2], ems[i,3]),output="address")
    zip[i] = str_extract(res, "\\d{5}")
}
```

```r
ems_coord_zip  <- cbind(data.frame((zip), stringsAsFactors=FALSE),ems)
```

```r
head(ems_coord_zip)
dim(ems_coord_zip)
```

```r
save(ems_coord_zip,file = 'ems_coord_zip.rds')
```
